```python
!pip3 install pickle5
import pickle5 as pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("./data/SGD.pickle", 'rb') as f:
SGD = pickle.load(f)
```

```python
with open('./data/ADAM.pickle', 'rb') as f:
    ADAM = pickle.load(f)
    print(len(ADAM))
```

```python
with open('./data/Dropout.pickle', 'rb') as f:
    Dropout = pickle.load(f)
    print(len(Dropout))
```

```python
with open('./data/BatchNorm.pickle', 'rb') as f:
    BatchNorm = pickle.load(f)
    print(len(BatchNorm))
```

```python
plt.figure(figsize=(15, 10))

plt.plot(np.array(SGD[17:]), 'g', label='SGD')
plt.plot(np.array(ADAM[17:]), 'r', label='ADAM')
plt.plot(np.array(Dropout[17:]), 'y', label='Dropout')
plt.plot(np.array(BatchNorm[17:]), 'k', label='BatchNorm')

plt.legend()
plt.grid('on')
plt.show()
```
